```python
# import sounddevice as sd
# import scipy.io.wavfile as wav
import numpy as np
# import plotly.express as px
# from scipy import signal
import librosa
import librosa.display
import matplotlib.pyplot as plt
plt.rcParams["figure.figsize"] = (10, 10)
```

```python
def plot_signal(wav_file):
    y, sr = librosa.load(wav_file)
    fig, ax = plt.subplots(nrows=2, ncols=1, sharex=False)

    ax[0].set(ylim=[-1, 1])
    librosa.display.waveshow(y, sr=sr, ax=ax[0])
    ax[0].xaxis.set_major_formatter(librosa.display.TimeFormatter(unit='s'))
    ax[0].set(xlabel='Time (s)')

    hop_length = 1024
    D = librosa.amplitude_to_db(np.abs(librosa.stft(y, hop_length=hop_length)), ref=np.max)
    img = librosa.display.specshow(D, y_axis='log', sr=sr, hop_length=hop_length, x_axis='time', ax=ax[1])
    fig.colorbar(img, ax=ax, format="%+2.f dB")
    ax[1].xaxis.set_major_formatter(librosa.display.TimeFormatter(unit='s'))
    ax[1].set(xlabel='Time (s)')
    
#plot_signal('../data/rec-2022-04-21/no-new-beetle3.wav')
```

```python
plot_signal('../data/rec-2022-04-21/no-new-beetle3.wav')
```
